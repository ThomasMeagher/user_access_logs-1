```python
import pandas as pd
emp = pd.read_csv('employees.csv')
emp.head()
```

```python
sys_access = pd.read_csv('system_access.csv')
sys_access.head()
```

```python
sys_access.rename(columns={'id':'employee_number'}, inplace=True)
sys_access['employee_number'] = 'EMP' + sys_access['employee_number'].astype(str).str.zfill(6)
sys_access.set_index('employee_number', inplace=True)
sys_access.head()
```

```python
emp.set_index('employee_number', inplace=True)
df = emp.join(sys_access)
df.head(10)
```

```python
update_df = df.loc[df['access_level'].isnull()]
update_df.head()
```

```python
update_df.loc[update_df['title'] == 'Director of Sales', 'access_level'] = 'Approval - Manager'
update_df.loc[update_df['title'] == 'Account Manager', 'access_level'] = 'Account Admin'
update_df.loc[update_df['title'] == 'Sales Assistant', 'access_level'] = 'Contract Entry'
update_df.head(10)
```

```python
update_df.to_csv('results.csv')
```
